```python
# dataframe as subsets

marketing_data = marketing_data[['ID','Year_Birth','Education',
                                 'Marital_Status','Income','Kidhome', 'Teenhome', 
                                 'Dt_Customer','Recency','NumStorePurchases',
                                 'NumWebVisitsMonth']]
marketing_data
```

```python

# group the no of store purchase based on kids at home 

list = marketing_data.groupby('Kidhome')['NumStorePurchases'].sum()
list
```

```python

# group the no of visits / month based on kids at home 

visits = marketing_data.groupby('Kidhome')['NumWebVisitsMonth'].sum()
visits
```
