```python
climate_temp = pd.read_csv('data/climate_temp.csv')
climate_percip = pd.read_csv('data/climate_precip.csv')
```

```python
# inner join

# first select the small slice of the dataset / focus on what we need

# and merge into dataset

percip_specific_station = climate_percip.query("STATION == 'GHCND:USC00045721'")
print("\n data",percip_specific_station)
print("s\n shape \n",percip_specific_station.shape)


# inner join

inner_merged = pd.merge(percip_specific_station, climate_temp)
```

```python
inner_merged


# inner join will discard the rows which doesn't match
```

```python
# we can also choose on which attribute, merge has to happen, we can pass more than one attribute

# inner join is specifies with 'on' parameter

inner_merged_total = pd.merge(climate_temp,climate_percip, on = ['STATION','DATE'])
inner_merged_total
```

```python


# outer join


# outer join is specifies with the 'how' paramter

# if a row doesn't have a match in other datframe based on the key values / columns , data will loose in inner join

# but it won't get lost in the outer join, data will be added with Nan values


```

```python
outer_merged = pd.merge(percip_specific_station,climate_temp, how='outer' , on=['STATION','DATE'])
```

```python
outer_merged
```

```python
climate_temp.shape
```

```python


# left outer join

# leave all the data from the left dataset, discard the data from the right dataset if the condition doesn't match


left_merged = pd.merge(climate_temp, percip_specific_station, how="left", on=['STATION','DATE'])
left_merged
```

```python
# if we change the order, it will result in different 

try_left_merged = pd.merge(percip_specific_station,climate_temp, how="left", on=['STATION','DATE'])
try_left_merged
```

```python
# right outer join is the mirror of left outer join

right_merged = pd.merge(percip_specific_station,climate_temp, how="right", on=['STATION','DATE'])
right_merged
```

```python
# pd.concat([climate_temp, climate_percip])  vs pd.concat([climate_temp, climate_percip], join="inner")


# The difference between pd.concat([climate_temp, climate_percip]) 
# and pd.concat([climate_temp, climate_percip], join="inner") lies in how they handle the concatenation of DataFrames 
# and the resulting shape of the output DataFrame.
```

```python
# pd.concat([climate_temp, climate_percip]):

# This performs an outer join by default. 
# It concatenates the DataFrames climate_temp and climate_percip along the specified axis (usually rows), 
# keeping all columns from both DataFrames.
# The resulting DataFrame will contain all rows from both DataFrames, 
# and missing values will be filled with NaN where data is absent in either DataFrame.
```

```python
# pd.concat([climate_temp, climate_percip], join="inner"):

# This performs an inner join. It concatenates the DataFrames climate_temp and climate_percip along the specified axis (usually rows), 
# but only includes columns that are present in both DataFrames.
# The resulting DataFrame will contain only rows where there are matching indices or rows in both DataFrames, 
# and it will include only columns that are common to both DataFrames.
```
