


**One Hot Vector Encoding**

```python
#list
B = ['bird', 'cat', 'dog', 'bird', 'bird']

# dictionary with one key category
d = {'categorical': B}
d
```

```python
# craete a new data frame based on the dict
df = pd.DataFrame(d)
df
```

```python
df['categorical']
```

```python
# dummies is used to create columns corresponding to unique values
dfDummies = pd.get_dummies(df['categorical'], prefix='category')
```

```python
dfDummies
```

```python
# we want vectors, not Dataframes

dfDummies.values[3]
```

```python
# Sample DataFrame
tempdata = {'categorical': ['A', 'B', 'A', 'C']}
tempdf = pd.DataFrame(tempdata)

# Create dummy variables
tempdfDummies = pd.get_dummies(df['categorical'], prefix='category')

# Get the values as a NumPy array
dummy_values = tempdfDummies.values.astype(bool)

print(dummy_values)
```
