```python
# covid data

data = pd.read_csv('data/Covid-cases-India.csv')
print(data)

# dropiing the s.no col

data.drop(data['S. No.'], axis=1, inplace=True)
```

```python
data.columns
```

```python
data.drop(data.columns[0], axis=1, inplace=True)
```

```python
data
```

```python
# adding the total cases

# adding india + foreign national cases

data['total_cases'] = data['Total Confirmed cases (Indian National)'] + data['Total Confirmed cases ( Foreign National )']
data['total_cases']
```

```python
data
```

```python
# total cases over all

total_cases_overall = data['total_cases'].sum()
total_cases_overall
```

```python
data['active_cases'] = data['total_cases'] - (data['Cured'] + data['Death'])
data['active_cases'] 
```
