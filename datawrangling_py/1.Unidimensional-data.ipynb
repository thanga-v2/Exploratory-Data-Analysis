```python
import matplotlib.pyplot as plt
```

```python

# heights a bell shaped distribution

plt.hist(heights, bins=13, color="lightgray", edgecolor="black")
plt.ylabel("Count")
plt.show()
```

```python
# The bell-shaped distribution is characterized by several key properties:

# Symmetry: The distribution is symmetric around its mean.
# Unimodality: There is a single peak or mode representing the most common value.
# Central tendency: The mean, median, and mode are all located at the same point.
# Dispersion: The spread of the distribution is characterized by the standard deviation.
# Probability density: The probability of observing values decreases as you move further away from the mean in either direction.
```

```python
# income - right skewed distribution


income = np.genfromtxt("https://raw.githubusercontent.com/gagolews/" +
    "teaching-data/master/marek/uk_income_simulated_2020.txt")
plt.hist(income, bins=20, color="lightgray", edgecolor="black")
plt.ylabel("Count")
plt.show()
```

```python
# In the context of income distribution, a right-skewed distribution would indicate that most individuals have lower incomes, 
# but there are a few individuals with very high incomes that pull the average (mean) income towards the higher end. 
# This pattern is commonly observed in income distributions, particularly in societies with significant income inequality.
```

```python
# bins
# change the bin limit accordingly to predict the bins
count,bins = np.histogram(income,25)
```
