```python
# 3 steps

# - split the original data and categorize into group
# - apply function to group
# - combine the result
```

```python
import numpy as np
import pandas as pd

data = pd.read_csv('data/complete.csv')
data
```

```python
data.shape

data.columns
```

```python
data = data[['awardYear', 'category','prizeAmount','prizeAmountAdjusted', 'name','gender','birth_continent',]]
data
```

```python
# step 1 - splitting into group

# lets split based on the category

grouped = data.groupby('category')
print(grouped)

grouped_category_year = data.groupby(['category','awardYear'])
print(grouped_category_year)
```

```python
grouped.indices
```

```python
grouped_category_year.indices
```

```python
# finding the number of groups in a group by object

print(grouped.ngroups)
print(grouped_category_year.ngroups)

print(len(grouped))
print(len(grouped_category_year))
```

```python
# If we need to visualize all or some entries of each group, we can iterate over the GroupBy object :

# grouped data has category as key and it has entries as values

for key,values in grouped:
    print(key)

for key,values in grouped_category_year:
    print('\n',key)
```
