```python
housing = pd.DataFrame({
    'rooms':[1,1,2,2,3,3,3],
    'price':[100,190,200,230,190,330,400]
})
```

```python
housing.T
```

```python
housing.style.background_gradient(cmap='Reds')
```

```python
# scatter plot



plt.scatter(x = housing['rooms'],y=housing['price']);


# continuous relationship betwwen rooms and price
```

```python
# style calls

plt.plot(housing['rooms'],housing['price']);
plt.title('Hotel rooms data')
plt.xlabel('Rooms')
plt.ylabel('Price')
```

```python
# style calls

plt.scatter(housing['rooms'],housing['price']);
plt.title('Hotel rooms data')
plt.xlabel('Rooms')
plt.ylabel('Price')
```

```python
plt.plot(housing['price'],housing['rooms'])
plt.title('without setting the limit')
```

```python
plt.plot(housing['price'],housing['rooms'])
plt.title('zooming the graph by setting x and y limit')
plt.xlabel('from 180 to 260')
plt.xlim(180,260)
plt.ylabel('from 2 to 3.5')
plt.ylim(2,3.5)
```

```python
plt.plot(housing['price'],housing['rooms'],color='red',marker='*',markersize=15)
plt.title('Marker')
```

```python
plt.plot(housing['price'],housing['rooms'],color='red',marker='o',markersize=15, linestyle='--')
plt.title('Marker and linestyle')
```

```python
import seaborn as sns
```

```python
data = pd.read_csv('data/heart.csv')
```

```python
data.head(5).T
```

```python



# Traget -> whether the person has any heart disease ?



```

```python


# continuos data vs categorical data

# state => categorical data -> male,female, pixel7, pixel7a GPT-3,GPT-4, versions etc..,

# no state => numerical data -> wright, height, kilometer, pulse etc


```

```python

# distribution plot

# it will be done on a continuos data

sns.distplot(data['age'])
```

```python
sns.displot(data['age'])
```

```python

# adding bins

sns.displot(data['age'],bins=50, color='red')
```

```python

# insight - people get heart disease around 40 to 70
# and it peaked at 60

# curve is called KDE - kernel density estimation

# it gives a general distribution
```

```python


# count plot


# it is for categorical values

```

```python

sns.countplot(data=data,x= 'sex')
```

```python
sns.countplot(data=data,x = 'slope')
```

```python
sns.countplot(data=data, x='target')
```

```python
sns.countplot(data=data, x=data['cp'])
```

```python



# hue in the count plot

# it will plot along with the column which we are mentioning


# if we want to plot the chest pain along with the gender

```

```python
sns.countplot(data=data, x= 'cp', hue='sex')
```

```python
sns.countplot(data=data, x= 'cp', hue='age')
```

```python




# box plot -> used to visualize between continuos and categorical data


# boxplot is a graph between continuos and categorical data
```

```python
# sns.boxplot( x - axis : categorical data, y - axis : continuos data) 
```

```python
sns.boxplot(x=data['sex'],y=data['age'])
```

```python


# if we want to check around target


sns.boxplot(x=data['target'], y=data['age'])

# 0 -> not having heart disease

# 1 -> having heart disease
```

```python
 # with respect to gender

sns.boxplot(x=data['target'], y=data['age'], hue=data['sex'])
```
